##***이진트리의 높이***

```python
class TNode : # 이진트리를 위한 노드 클래스
  def __init__ (self, data, left, right) : # 생성자
    self.data = data # 노드의 네이터
    self.left = left # 왼족 자식을 위한 링크
    self.right = right # 오른쪽 자신을 위한 링크

def calc_height(root) : # 루트노드의 root
  if root is None : # 공백트리이면 0 반환
    return 0 # 높이가 0 이면 0 반환
  hleft = calc_height(root.left) # 왼쪽 트리의 높이
  hright = calc_height(root.right) # 오른쪽 트리의 높이
  return max(hleft, hright) + 1 # 더 높은 높이에 1을 더하여 반환
  
d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c) # 루트 node
print("트리의 높이 = ", calc_height(root))
```

## ***표준 순회 알고리즘***
#### ***1. 이진트리의 전위 순회***
#### ***2. 이진트리의 중위 순회***
#### ***3. 이진트리의 후위 순회***

```python
# <1>
def preorder(n) : # 전위 순회 함수
  if n is not None : 
    print(n.data, end = ' ')
    preorder(n.left)
    preorder(n.right)

# <2>
def inorder(n) : # 중위 순회 함수
  if n is not None :
    inorder(n.left)
    print(n.data, end = ' ')
    inorder(n.right)

# <3>
def postorder(n) : # 후위 순회 함수
  if n is not None :
    postorder(n.left)
    postorder(n.right)
    print(n.data, end = ' ')

d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c) # 루트 node

print(' In-order : ', end = ' ')
inorder(root)
print()
print(' pre-order :', end = ' ' )
preorder(root)
print()
print(' post-order :', end = ' ' )
postorder(root)
print()
```
