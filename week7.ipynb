##***기수정렬(radix_sort)***

```python
from queue import Queue # python queue 모듈의 queue 사용
def radix_sort(A): 
    queues = [] # 큐의 리스트
    for i in range(BUCKETS):
        queues.append(Queue()) # 버킷 큐 사용
        
    n = len (A)
    factor = 1 # 1의 자리부터 시작
    for d in range(DIGITS): # 모든 자리에 대해
        for i in range(n): # 자릿수에 따라 큐 삽입
            queues[(A[i]//factor) % 10].put(A[i]) # 숫자를 삽입
        j = 0
        for b in range(BUCKETS): # 버킷에서 꺼내어 원래 리스트로
            while not queues[b].empty(): # b번째 큐가 공백이 아닌 
                A[j] = queues[b].get() # 원소를 꺼내 리스트에 저장
                j += 1
        factor *= 10
        print("STEP", d + 1, A)
        
import random
BUCKETS = 10
DIGITS = 4
data = []
for i in range(10):
    data.append(random.randint(1, 9999))
radix_sort(data)
print("Radix:", data)
```

##***계수정렬(counting_sort)***

```python
def counting_sort(A):
    output = [0] * len(A)
    count = [0] * MAX_VAL
    
    for i in A:
        count[i] += 1
        
    for i in range(MAX_VAL):
        count[i] += count[i - 1]
    
    for i in range(len(A)):
        output[count[A[i]] - 1] = A[i]
        count[A[i]] -= 1
        
    for i in range(len(A)):
        A[i] = output[i]
        
MAX_VAL = 10
data = [1, 4, 6, 1, 4 ,3 , 7, 3, 2]
print("ORIGINAL : ", data)
counting_sort(data)
print("COUNTING : ", data)
```
