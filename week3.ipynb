## ***위상정렬(축소정복 기법)***

```python
#위상정렬(축소정복기법)
import queue
def topological_sort(graph) :
  inDeg = {} #{정점:진입차수}저장을 위한
  for v in graph : #그래프의 모든 정점에 대해
    inDeg[v] = 0 #0으로 초기화, indeg={}
  for v in graph : #모든 정점v에 대해
    for u in graph[v] : #v에 인접한 모든 정점 u에 대해
      inDeg[u] += 1 # 진입차수를 1 증가시킴
  print(inDeg)
  vlist = [] # 진입차수가0인 정점 리ㅡ트를
  for v in graph : 
    if inDeg[v] == 0 : #진입차수가 0이면 vlist에 추가
      vlist.append(v)
  
  while vlist : #리스트가 공백이 아닐 때 까지
    v = vlist.pop() #진입차수가 0인 정점을 하나 
    print(v, end=' ') #화면 출력

    for u in graph[v] : #연결된 정점의 진입차수 감소
      inDeg[u] -= 1 #진입차수 감소
      if inDeg[u] == 0 : #지입차수가 0만 이면 vlist에 추가
        vlist.append(u)
        
############################################

def topological_sort2(graph) :
  inDeg = {} 
  for v in graph : 
    inDeg[v] = 0 
  for v in graph : 
    for u in graph[v] : 
      inDeg[u] += 1 
  #print(inDeg)
  #vlist = []
  vlist = queue.Queue()
  for v in graph : 
    if inDeg[v] == 0 : 
      vlist.put(v)
  
  while not vlist.empty : 
    v = vlist.get() 
    print(v, end=' ')

    for u in graph[v] :
      inDeg[u] -= 1 
      if inDeg[u] == 0 :
        vlist.put(u)

mygraph = {"A" : {"C", "D"},
           "B" : {"D", "F"},
           "C" : {"D", "F"},
           "D" : {"F"},
           "E" : {"F"},
           "F" : set()
           }
print("topological_sort: ")
topological_sort(mygraph)
print()
topological_sort2(mygraph)
print()
```

```python
import queue
def bfs(graph, start) :
  visited = { start }
  que = queue.Queue()
  que.put(start)
  while not que.empty() : 
    v = que.get()
    print(v, end= ' ')
    nbr = graph[v] - visited
    for u in nbr :
      visited.add(u)
      que.put(u)
    

mygraph = {"A" : {"C", "D"},
           "B" : {"D", "F"},
           "C" : {"D", "F"},
           "D" : {"F"},
           "E" : {"F"},
           "F" : set()
           }
print('BFS : ', end = '')
bfs(mygraph, "A")
print()
```
