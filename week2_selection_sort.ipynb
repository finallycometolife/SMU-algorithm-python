## ***선택정렬***



```python
def selection_sort(A) : #선택정렬
    n = len(A) #n: 리스트 A의 길이
    for i in range(n-1) : #0, 1, 2 ---- n-2 외부 loop
        least = i
        for j in range(i+1, n) : #i+1, ---- n-1 내부 loop
            if (A[j]<A[least]) : # 비교연산
                least = j #최소항목 갱신
        A[i], A[least] = A[least], A[i] #배열 항목 교환
        printStep(A, i + 1); # 중간 과정 출력

def printStep(arr, val) : # 중간 과정 출력 f(x)
    print("step %2d = " % val, end= '') # 현재 단계를 출력
    print(arr) # 리스트 내용 출력
    
data = [ 5, 3, 8, 4, 9, 1, 6, 2, 7]
print("original : ", data)
selection_sort(data)
print("selection :", data)
```

## ***순차 탐색***

```python
#순차 탐색
def sequential_search(A, key): # A는 입력 리스트 key는 탐색키
  for i in range(len(A)): # i: 0, 1----- len(A)-1
    if A[i] == key: # 탐색 성공하면 
      return i # 인덱스 반환
    return -1 # 실패시 -1 반환
  
A = [ 32, 14, 5, 17, 23, 9, 11, 4, 26, 29]
print("find 32: ", sequential_search(A, 32))
print("find 26: ", sequential_search(A, 32))
```

## ***억지 기법***

```python
#문자열 매칭(억지 기법)
def string_matching(T, P): # T는 입력, P는 패턴
  n = len(T) # n은 입력 문자열의 길이
  m = len(P) # m은 패턴 문자열의 길이
  for i in range(n-m+1): # i: 0, 1 -----, n-m
    j = 0
    while j < m and P[j]==T[i+j]: # 모든 문자가 일치하면 매칭 성공
      j = j + 1 # 현재 위치 반환
    if j == m: # 문자열 매칭 fail
      return i
  return -1

text = 'HELLO WORLD'
pattern = 'HEL'
print(pattern, 'in', text, '-->', string_matching(text, pattern))
pattern = 'Hi'
print(pattern, 'in', text, '-->', string_matching(text, pattern))
```

## ***최근접 점의 쌍의 거리 구하기***

```python
#최근접 점의 쌍의 거리
import math # math 모듈 포함, 제곱근(sqrt) 함수
def distance(p1, p2): # euclidean 거리 계산 함수
  return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

def closet_pair(p):
  n = len(p) # 점의 전체 개수
  mindist = float("inf") # 최근점 거리 초기화(무한)
  for i in range(n-1):
    for j in range(i+1, n):
      dist = distance(p[i], p[j]) # 유클리드 거리 계산
      if dist < mindist:
        mindist = dist
        ninp1 = i
        minp2 = j
  return mindist, ninp1, minp2

p = [(55, 66), (12, 30), (40, 50), (1, 2), (12, 10), (3, 9)]
print("최근접 거리:", closet_pair(p))
```

## ***깊이 우선 탐색***

```python
#깊이 우선 탐색
def dfs(graph, start, visited ):
  if start not in visited: # start가 방문하지 않은 정점이면
    visited.add(start) # start를 방문한 노드 집합에 추가
    print(start, end= ' ') # start를 방문했다고 출력
    nbr = graph[start] - visited # nbr은 차집합 연산 이용
    for v in nbr: # v는 {인접정점} - {방문정점}에 포함됨
      dfs(graph, v, visited) # v에 대해 dfs를 순환적 호출

mygraph = { "A" : {"B", "C"},
           "B" : {"A", "D"},
           "C" : {"A", "D", "E"},
           "D" : {"B", "C", "F"},
           "E" : {"C", "G", "H"},
           "F" : {"D"},
           "G" : {"E", "H"},
           "H" : {"E", "G"}
           }
print('DFS: ', end = '')
dfs(mygraph, "A", set())
print()
```

## ***너비 우선 탐색***

```python
#너비 우선 탐색
import queue
def bfs(graph, start):
  visited = {start} # 맨 처음에는 start만 방문한 정점
  que = queue.Queue() # 파이썬 큐 모듈의 큐 객체 삽입
  que.put(start)
  while not que.empty(): # 큐에 항목이 있을 때 까지
    v = que.get() # 큐에서 하나의 정점 v를 추출
    print(v, end = '') # v는 방문했음을 출력
    nbr = graph[v] - visited # nbr = (v의 인접정점) - (방문정점)
    for u in nbr: # 갈 수 있는 모든 인접 정점에 대해
      visited.add(u) # u는 방문 해씀
      que.put(u) # u를 큐에 삽입

mygraph = { "A" : {"B", "C"},
           "B" : {"A", "D"},
           "C" : {"A", "D", "E"},
           "D" : {"B", "C", "F"},
           "E" : {"C", "G", "H"},
           "F" : {"D"},
           "G" : {"E", "H"},
           "H" : {"E", "G"}
          }

print('BFS: ', end = '')
bfs(mygraph, "A")
print()
```
